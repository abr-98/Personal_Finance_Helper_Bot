```python
data_filtered["Total_investment_options"] = data_filtered["Mutual_Funds"] + data_filtered["Equity_Market"] + data_filtered["Debentures"] + data_filtered["Government_Bonds"] + data_filtered["Fixed_Deposits"] + data_filtered["PPF"] + data_filtered["Gold"]
```

```python
### divide each investment option by total investment options to get the percentage of each investment option

data_filtered["Mutual_Funds_Percentage"] = data_filtered["Mutual_Funds"] / data_filtered["Total_investment_options"]
data_filtered["Equity_Market_Percentage"] = data_filtered["Equity_Market"] / data_filtered["Total_investment_options"]
data_filtered["Debentures_Percentage"] = data_filtered["Debentures"] / data_filtered["Total_investment_options"]
data_filtered["Government_Bonds_Percentage"] = data_filtered["Government_Bonds"] / data_filtered["Total_investment_options"]
data_filtered["Fixed_Deposits_Percentage"] = data_filtered["Fixed_Deposits"] / data_filtered["Total_investment_options"]
data_filtered["PPF_Percentage"] = data_filtered["PPF"] / data_filtered["Total_investment_options"]
data_filtered["Gold_Percentage"] = data_filtered["Gold"] / data_filtered["Total_investment_options"]
data_filtered.head()
```

```python
### Clustering based on gender, age, objective, purpose, and expectation
```

```python
data_filtered.Objective.value_counts()
```

```python
### gender, Objective, Purpose, Expect encoding

from sklearn.preprocessing import LabelEncoder
import pickle

for column in ["gender", "Objective", "Purpose", "Expect"]:
    le = LabelEncoder()
    data_filtered[column] = le.fit_transform(data_filtered[column])
    with open(f"{column}_label_encoder.pkl", "wb") as f:
        pickle.dump(le, f)

```

```python
### grouping dataset 

data_grouped = data_filtered[["gender", "Objective", "Purpose", "Expect","age"]]
```

```python
data_grouped.head()
```

```python
from sklearn.cluster import KMeans

## try elbow method to find the optimal number of clusters

inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_grouped)
    inertia.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1, 11), inertia, marker="o")
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data_grouped)
```

```python
import pickle
```

```python
with open("cluster.pkl", "wb") as f:
    pickle.dump(kmeans, f)
```

```python
data_grouped["cluster"] = kmeans.labels_
```

```python
data_grouped.head()
```

```python
data_filtered.drop(columns=["gender","Objective","Purpose", "Expect", "age"], inplace=True)
```

```python
data_final = pd.concat([data_filtered, data_grouped], axis=1)
```

```python
data_final.head()
```

```python
mean_investment_options = data_final.groupby("cluster")[["Mutual_Funds_Percentage", "Equity_Market_Percentage", "Government_Bonds_Percentage", "Debentures_Percentage","Gold_Percentage", "PPF_Percentage", "Fixed_Deposits_Percentage"]].mean()
```
