```python
data_filtered["Total_investment_options"] = data_filtered["Mutual_Funds"] + data_filtered["Equity_Market"] + data_filtered["Debentures"] + data_filtered["Government_Bonds"] + data_filtered["Fixed_Deposits"] + data_filtered["PPF"] + data_filtered["Gold"]
```

```python
### divide each investment option by total investment options to get the percentage of each investment option

data_filtered["Mutual_Funds_Percentage"] = data_filtered["Mutual_Funds"] / data_filtered["Total_investment_options"]
data_filtered["Equity_Market_Percentage"] = data_filtered["Equity_Market"] / data_filtered["Total_investment_options"]
data_filtered["Debentures_Percentage"] = data_filtered["Debentures"] / data_filtered["Total_investment_options"]
data_filtered["Government_Bonds_Percentage"] = data_filtered["Government_Bonds"] / data_filtered["Total_investment_options"]
data_filtered["Fixed_Deposits_Percentage"] = data_filtered["Fixed_Deposits"] / data_filtered["Total_investment_options"]
data_filtered["PPF_Percentage"] = data_filtered["PPF"] / data_filtered["Total_investment_options"]
data_filtered["Gold_Percentage"] = data_filtered["Gold"] / data_filtered["Total_investment_options"]
data_filtered.head()
```
